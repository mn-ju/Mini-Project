```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
datadir = '/content/drive/MyDrive/data'
```

```python
categories = ['cow','fish'] # Name the folder with these categories
target = []
flat_data = []
for i in categories:
  path = os.path.join(datadir,i)
  #print(path)
  for img in os.listdir(path):
    print(img)
    img_array = imread(os.path.join(path,img))

    img_resize = resize(img_array,(50,50,3)) # Value gets converted 0-1 from 0-255
   
    flat_data.append(np.ndarray.flatten(img_resize))
    target.append(i)
    plt.imshow(img_resize)
    plt.show()

flat_data = np.array(flat_data)
target = np.array(target)
```

```python
df = pd.DataFrame(flat_data)
df['Target'] = target
df
```

```python
np.unique(target,return_counts=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data,target,random_state=0)
```

## **Support Vector Machine**

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100],
              'kernel':['linear','rbf']}

grid = GridSearchCV(SVC(),param_grid)
grid.fit(x_train,y_train)
grid.best_params_
```

```python
from sklearn.svm import SVC
model = SVC(C=10)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
confusion_matrix(y_pred,y_test)
accuracy_score(y_pred,y_test)
```
